```python
# ЗАДАЧА Реализовать функцию, осуществляющую матричные операции для получения theta
def linreg_linear(X, y):
    Xt=X.transpose()
    theta=np.linalg.inv(Xt.dot(X)).dot(Xt).dot(y)
    return theta
```
